# Caso de estudo

Um dos problemas mais comums de classificação é a identificação de flores. Um exemplo disso se tornou muito famoso, é o caso da classificação das flores Iris.

Trata-se de uma classificação multi-dimensional, onde um conjunto de 4 características da flor conseguem definir a classe da planta.

## A Espécie _Iris_

Essa espécie se tornou muito famosa no meio científico por sua "utilidade estatística". Em 1936 Robert Fisher publicou um estudo taxonômico desta planta baseada em 4 critérios métricos _[1]_:         
 - Comprimento da **Pétala**
 - Largura da **Pétala**
 - Comprimento da **Sépala**
 - Largura da **Sépala**

## Classificação Morfológica

Posteriormente outro pesquisador publicou uma classificação do dataset publicado por Fisher em 3 classes de plantas:

- Setosa
- Versicolor
- Virgínica

<img src='https://miro.medium.com/max/1000/1*Hh53mOF4Xy4eORjLilKOwA.png' width='100%'/>

A série de dados criada por Fisher se tornou tão famosa que passou a ser utilizada como benchmark para modelos de classificação.

O __Iris Flower Data Set__ é provavelmente a base de dados mais bem conhecida da literatura de reconhecimento de padrões.

- 3 classes com 50 indivíduos em cada.
- Cada classe representa um tipo de planta.
- A primeira classe é linearmente independente das demais.
- As outras duas classes não são linearmente independentes entre si.

# Objetivo

O objetivo principal do trabalho então é utilizar o algorítmo de otimização do Lobo Cinzento (GWO) para obter uma classificação que se aproxime da classificação definida no dataset.

Isso é possível já que o problema de classificação pode ser compreendido como um problema de otimização iterativa onde procura-se obter grupos que **maximizem a variância entre si**, mas que **minimizem a variância entre os elementros dentro do grupo**.
